```python
from pyiron_base import Project, job
```

```python
@job
def get_sum(x, y):
    return x + y
```

```python
pr = Project("test")
pr.remove_jobs(recursive=True, silently=True)
```

```python
result_1 = get_sum(x=1, y=2, pyiron_project=pr)
result_1.pull()
```

```python
result_2 = get_sum(x=2, y=3, pyiron_project=pr)
result_2.pull()
```

```python
df = pr.job_table()
df
```

```python
def get_input(job):
    return job.content["storage"]["input"].to_builtin()
```

```python
def get_output(job):
    return job.content["storage"]["output"].to_builtin()
```

```python
table = pr.create.table()
table.add["function_input"] = get_input
table.add["function_output"] = get_output
table.run()
result = table.get_dataframe()
result
```
