```python
from pyiron_base import Project, job
```

```python
@job(output_key_lst=['prod', 'div'])
def get_prod_and_div(x, y):
    return {"prod": x * y, "div": x / y}
```

```python
@job
def get_sum(x, y):
    return x + y
```

```python
@job
def get_square(x):
    return x ** 2
```

```python
@job
def get_inner_part(a, b):
    pr = Project(".")
    prod_and_div = get_prod_and_div(x=1, y=2, pyiron_project=pr)
    return get_sum(x=prod_and_div.output.prod, y=prod_and_div.output.div, pyiron_project=pr).pull()
```

```python
@job
def get_total_workflow(c, d):
    pr = Project(".")
    tmp_sum = get_inner_part(a=c, b=d, pyiron_project=pr)
    tmp_sum.execute_in_working_directory = True
    result = get_square(x=tmp_sum, pyiron_project=pr)
    result.execute_in_working_directory = True
    return result.pull()
```

```python
pr = Project("test")
pr.remove_jobs(recursive=True, silently=True)
```

```python
result = get_total_workflow(c=1, d=2, pyiron_project=pr)
result.execute_in_working_directory = True
```
